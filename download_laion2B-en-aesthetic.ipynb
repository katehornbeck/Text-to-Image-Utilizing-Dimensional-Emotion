```python
import requests
import os
import pandas

# Headers to avoid 403 Forbidden errors
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Function to download and save an image
def download_image(url, output_path):
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Check for request errors
        with open(output_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {output_path}")
    except Exception as e:
        print(f"Failed to download {url}: {e}")

def construct_dataset(folder, num):
    output_dir = folder + 'part-' + str(num).zfill(5) + "-images"
    os.makedirs(output_dir, exist_ok=True)

    # find the parquet file    
    parquet_file = [ x for x in os.listdir(folder) if x.endswith('parquet') and x.startswith('part-'+str(num).zfill(5))][0]
    print(f'Working on {parquet_file}')
    df = pandas.read_parquet(folder + parquet_file)

    # save info needed for csv
    texts = df[["URL","TEXT"]]
    texts.to_csv(output_dir + "/part-" + str(num).zfill(5) + ".csv")

    # Iterate through the column and download images
    for index, url in df["URL"].items():
        print(url)
        if isinstance(url, str) and url.startswith('http'):  # Check if the URL is a string and not a file path
            # Generate a file name based on the index or URL
            file_name = f"image_{index}.jpg"
            output_path = os.path.join(output_dir, file_name)
            download_image(url, output_path)
```

```python
for i in range(3):
    construct_dataset("../datasets/laion2B-en-aesthetic/", i)
```
